## Loading a Model

```python
import irtk

# Load pretrained (GPT-2, GPT-Neo, Pythia, etc.)
model = irtk.HookedTransformer.from_pretrained("gpt2")  # or "gpt2-medium", "pythia-70m", etc.

# Create from config (random weights)
cfg = irtk.HookedTransformerConfig(
    n_layers=2, d_model=64, n_ctx=128, d_head=16, n_heads=4, d_vocab=1000,
)
model = irtk.HookedTransformer(cfg)

# Access config
model.cfg.n_layers, model.cfg.d_model, model.cfg.n_heads
```

## Forward Pass & Caching

```python
import jax.numpy as jnp

tokens = jnp.array([50256, 15496, 995])  # [BOS, Hello, world]

# Basic forward
logits = model(tokens)                          # [seq_len, d_vocab]

# Forward + cache all activations
logits, cache = model.run_with_cache(tokens)    # cache: ActivationCache

# Forward with hook interventions
def my_hook(x, name):
    return x * 0  # zero ablation
logits = model.run_with_hooks(tokens, fwd_hooks=[("blocks.0.hook_attn_out", my_hook)])
```

## ActivationCache

```python
logits, cache = model.run_with_cache(tokens)

# Access activations
cache["blocks.0.attn.hook_q"]              # full name
cache[("q", 0)]                             # shorthand: (act_name, layer)
cache[("pattern", 2)]                       # attention pattern at layer 2
cache[("resid_post", 3)]                    # residual stream after layer 3

# Available hooks per layer:
# hook_resid_pre, hook_attn_out, hook_resid_mid, hook_mlp_out, hook_resid_post
# attn: hook_q, hook_k, hook_v, hook_z, hook_attn_scores, hook_pattern, hook_result
# mlp: hook_pre, hook_post
# ln: hook_scale, hook_normalized

# Residual stream analysis
resid_stack = cache.accumulated_resid()                    # [n_layers+1, seq, d_model]
components = cache.decompose_resid()                       # individual contributions
head_results = cache.stack_head_results(layer=0)           # [n_heads, seq, d_model]
all_q = cache.stack_activation("q")                        # [n_layers, seq, n_heads, d_head]

# Logit attribution
logit_attrs = cache.logit_attrs(resid_stack)               # [n_components, seq, d_vocab]
logit_attrs = cache.logit_attrs(resid_stack, tokens=tokens)  # [n_components, seq] per-token
```

## Activation Patching

```python
from irtk.patching import (
    activation_patch, patch_by_layer, patch_by_head, patch_by_position,
    zero_ablate, mean_ablate, ablate_heads,
    path_patch, path_patch_by_receiver,
    make_logit_diff_metric, make_loss_metric,
)

metric = make_logit_diff_metric(correct_token=42, wrong_token=13)

# Patch entire activations
results = activation_patch(model, clean, corrupted, ["blocks.0.hook_resid_post"], metric)
layer_results = patch_by_layer(model, clean, corrupted, metric)    # [n_layers]
head_results = patch_by_head(model, clean, corrupted, metric)      # [n_layers, n_heads]
pos_results = patch_by_position(model, clean, corrupted, metric)   # [n_layers, seq_len]

# Ablation
zero_results = zero_ablate(model, tokens, ["blocks.0.hook_attn_out"], metric)
mean_results = mean_ablate(model, tokens, ["blocks.0.hook_attn_out"], metric)
head_ablation = ablate_heads(model, tokens, metric, method="zero")  # [n_layers, n_heads]

# Path patching (fine-grained)
result = path_patch(model, clean, corrupted,
    sender_hook="blocks.0.hook_attn_out",
    receiver_hooks=["blocks.1.attn.hook_q"],
    metric_fn=metric)
receiver_results = path_patch_by_receiver(model, clean, corrupted,
    sender_hook="blocks.0.hook_attn_out", metric_fn=metric)
```

## FactoredMatrix

```python
from irtk import FactoredMatrix

fm = FactoredMatrix(A, B)     # represents A @ B without materializing
fm.shape                       # shape of the full product

fm.BA                          # B @ A (useful when small)
fm.T                           # transpose
U, S, Vh = fm.svd()           # efficient SVD
eigenvals = fm.eigenvalues     # eigenvalues (square matrices)
norm = fm.norm()               # Frobenius norm
result = fm @ other_fm         # compose FactoredMatrices
```

## Common Patterns

### Get all hook point names
```python
print(list(model.hook_dict.keys()))
```

### Cache specific hooks only (faster)
```python
from irtk.hook_points import HookState
cache = {}
hs = HookState(hook_fns={}, cache=cache)
# Only hooks that exist in the model will cache
logits = model(tokens, hook_state=hs)
```

### JIT compilation (no hooks)
```python
import equinox as eqx
fast_model = eqx.filter_jit(model)
logits = fast_model(tokens)  # much faster, no hook overhead
```

### Batch processing
```python
import jax
batched_logits = jax.vmap(model)(batch_of_tokens)  # [batch, seq, vocab]
```

## Gradient-Based Interpretability

```python
from irtk.gradients import (
```
