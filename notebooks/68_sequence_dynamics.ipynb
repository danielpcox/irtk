# Sequence DynamicsHow do transformers process sequence structure? Analyzes repetition handling, long-range dependencies, position bias, length effects, and boundary effects.

## Why This Matters

Sequence dynamics analysis reveals how models handle positional patterns: repetition detection, long-range dependencies, position bias, and length effects. Understanding these dynamics is crucial for explaining why models succeed or fail on specific sequence structures.

**Key references:**
- [Olsson et al. (2022) "In-context Learning and Induction Heads"](https://arxiv.org/abs/2209.11895) — Discovers induction heads and training phase transitions
- [Elhage et al. (2021) "A Mathematical Framework for Transformer Circuits"](https://transformer-circuits.pub/2021/framework/index.html) — Foundational framework for attention head and residual stream analysis

```python
import jaximport jax.numpy as jnpimport numpy as npfrom irtk import HookedTransformerConfig, HookedTransformercfg = HookedTransformerConfig(n_layers=2, d_model=16, n_ctx=32, d_head=4, n_heads=4, d_vocab=50)model = HookedTransformer(cfg)key = jax.random.PRNGKey(42)leaves, treedef = jax.tree.flatten(model)new_leaves = []for leaf in leaves:    if isinstance(leaf, jnp.ndarray) and leaf.dtype == jnp.float32:        key, sk = jax.random.split(key)        new_leaves.append(jax.random.normal(sk, leaf.shape, dtype=leaf.dtype) * 0.1)    else:        new_leaves.append(leaf)model = jax.tree.unflatten(treedef, new_leaves)tokens = jnp.array([0, 1, 2, 3, 4, 5, 6, 7])def metric(logits):    return float(logits[-1, 0])
```

```python
from irtk.sequence_dynamics import (    repetition_handling_analysis,    long_range_dependency_tracking,    position_bias_strength,    length_effect_on_circuits,    boundary_effect_analysis,)
```

## Repetition HandlingDoes the model detect repeated patterns via induction or just memorize tokens?

```python
# Sequence with repetitions: [A B C A B C]rep_tokens = jnp.array([5, 10, 15, 5, 10, 15])result = repetition_handling_analysis(model, rep_tokens)print(f"First occurrence entropy: {result['first_occurrence_entropy']:.4f}")print(f"Repeated occurrence entropy: {result['repeated_occurrence_entropy']:.4f}")print(f"Entropy reduction: {result['entropy_reduction']:.4f}")print(f"Induction score: {result['induction_score']:.4f}")print(f"Repeated tokens: {result['n_repeated_tokens']}")
```

## Long-Range Dependency TrackingHow well does information flow from a distant source to the target position?

```python
result = long_range_dependency_tracking(model, tokens, source_pos=0, target_pos=-1)print(f"Direct attention per layer: {[f'{a:.4f}' for a in result['direct_attention_per_layer']]}")print(f"Indirect attention: {result['indirect_attention_estimate']:.6f}")print(f"Source ablation effect: {result['source_ablation_effect']:.6f}")print(f"Information retention: {result['information_retention']:.4f}")print(f"Effective distance: {result['effective_distance']:.2f}")
```

## Position BiasDoes the model have systematic recency or primacy biases?

```python
result = position_bias_strength(model, tokens)print(f"Recency bias: {result['recency_bias']:.4f}")print(f"Primacy bias: {result['primacy_bias']:.4f}")print(f"Distance decay rate: {result['distance_decay_rate']:.4f}")print(f"Position entropy: {result['position_entropy']:.4f}")
```

## Length Effects on CircuitsHow does circuit behavior change with sequence length?

```python
result = length_effect_on_circuits(model, tokens, metric, lengths=[2, 4, 6, 8])for l, m in zip(result['lengths'], result['metrics']):    print(f"  Length {l}: metric = {m:.6f}")print(f"Length sensitivity: {result['length_sensitivity']:.6f}")print(f"Metric variance: {result['metric_variance']:.8f}")print(f"Stable beyond length: {result['stable_beyond']}")
```

## Boundary EffectsHow do sequence edges (start/end) affect computation?

```python
result = boundary_effect_analysis(model, tokens)print(f"Start norm ratio: {result['start_norm_ratio']:.4f}")print(f"End norm ratio: {result['end_norm_ratio']:.4f}")print(f"Start attention concentration: {result['start_attention_concentration']:.4f}")print(f"End confidence boost: {result['end_confidence_boost']:.4f}")print(f"Boundary effect strength: {result['boundary_effect_strength']:.4f}")
```
